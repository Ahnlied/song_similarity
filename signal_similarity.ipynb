```python
songs_signals = [to_plot_binks_jap,
                to_plot_binks_spa,
                to_plot_peaches,
                to_plot_fast_car,
                to_plot_sunshine,
                to_plot_crying,
                to_plot_castigo]
```

```python
songs_names = ['Binks jap','Binks spa','Peaches','Fast Car','Sunshine','Crying','El Castigo']
```

```python
from fastdtw import fastdtw

confussion = np.zeros((len(songs_signals), len(songs_signals)))

for i in range(0,len(songs_signals)):
    song_1 = songs_signals[i]
    print(len(song_1))
    for j in range(0,len(songs_signals)):
        song_2 = songs_signals[j]
        d, path = fastdtw(song_1,song_2)
        confussion[i,j] = d

confussion
#d, path = fastdtw(a,b)
```

```python
fig, ax = plt.subplots()
im = ax.imshow(confussion)

# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(songs_signals)), labels=songs_names)
ax.set_yticks(np.arange(len(songs_signals)), labels=songs_names)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(0,len(songs_names)):
    for j in range(0,len(songs_names)):
        text = ax.text(j, i, confussion[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")
fig.tight_layout()
plt.show()
```
